### Splitting X_train, Y_train, X_test and Y_test

```python
X_train = loan_df.iloc[:, :-1].values
Y_train= loan_df.iloc[:, -1].values.reshape(-1,1)
```

```python
X_train.shape, Y_train.shape
```

```python
X_test = test_df.iloc[:, :-1].values
Y_test= test_df.iloc[:, -1].values.reshape(-1,1)
```

```python
X_test.shape, Y_test.shape
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model1 = GradientBoostingClassifier(n_estimators=30, learning_rate=1, max_features=5, max_depth=5, random_state=0)
model1.fit(X_train, Y_train.ravel())
Y_pred=model1.predict(X_test)
from sklearn.metrics import precision_score, recall_score
ps=precision_score(Y_test,Y_pred)
rs=recall_score(Y_test,Y_pred)
print("Accuracy score :",model1.score(X_test,Y_test))
print("Precision: ",ps)
print("Recall: ",rs)
```

### Let us analyse the different accuracies for increasing number of trees:

```python
est_list=[10,20,30,40,50]
for n_est in est_list:
    model2=GradientBoostingClassifier(n_estimators=n_est, learning_rate=1, max_features=5, max_depth=5, random_state=0)
    model2.fit(X_train, Y_train.ravel())
    Y_pred=model2.predict(X_test)
    ps=precision_score(Y_test,Y_pred)
    rs=recall_score(Y_test,Y_pred)
    print("The accuracy for",n_est,"estimators with learning rate=1, max_features=5 and max_depth=5 is",model2.score(X_test,Y_test),"Precision: ",ps,"Recall: ",rs)
```

#### Therefore the accuracy increases in most of the cases when the number of trees increases

Let us analyse the different accuracies for different learning rate:

```python
#let us fix the n_estimators as 30 as it gave the second highest accuracy
lr_list = [0.05, 0.075, 0.1, 0.25, 0.5, 0.75, 1]
for lr in lr_list:
    model3=GradientBoostingClassifier(n_estimators=30, learning_rate=lr, max_features=5, max_depth=5, random_state=0)
    model3.fit(X_train, Y_train.ravel())
    Y_pred=model3.predict(X_test)
    ps=precision_score(Y_test,Y_pred)
    rs=recall_score(Y_test,Y_pred)
    print("The accuracy for 30 estimators with learning rate=",lr,",max_features=5 and max_depth=5 is",model3.score(X_test,Y_test),"Precision: ",ps,"Recall: ",rs)
    
```

Therefore, the accuracy increases with increase in learning rate

```python
#let us fix n_estimators=30 and learning_rate=0.5
l=[1,2,3,4,5]
for i in l:
    model4=GradientBoostingClassifier(n_estimators=30, learning_rate=0.5, max_features=i, max_depth=i, random_state=0)
    model4.fit(X_train, Y_train.ravel())
    model4.fit(X_train, Y_train.ravel())
    Y_pred=model4.predict(X_test)
    ps=precision_score(Y_test,Y_pred)
    rs=recall_score(Y_test,Y_pred)
    print("The accuracy for 30 estimators with learning rate= 0.5, max_features=",i, "and max_depth=",i,"is",model4.score(X_test,Y_test),"Precision: ",ps,"Recall: ",rs)
```

Hence the best test accuracy is obtained when there are 30 estimators, learning rate=0.5, max_features=5 and max_depth=5 which is 0.9949565704679182

### Comparing with simple decision tree

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier(criterion='entropy',max_features=5, max_depth=5)
dtree.fit(X_train,Y_train)
Y_pred=dtree.predict(X_test)
ps=precision_score(Y_test,Y_pred)
rs=recall_score(Y_test,Y_pred)
acc=dtree.score(X_test,Y_test)
print("Accuracy: ",acc,"Precision: ",ps,"Recall: ",rs)
```

Therefore, from the above accuracies, we can see that the Gradient boosting (accuracy:0.9949565704679182) performs much better than the decision tree (accuracy: 0.9618240403474363)
